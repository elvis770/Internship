EXTRACTING THE MARKERS AND SPLITTING THE DATASET INTO 10 DIFFERENT BATCHES DUE TO COMPUTING RESOURCES 

```python
import os
import pandas as pd
import json
import pyvista as pv
import h5py
import base64

def extract_markers_parameters(txt_path):
    parameters = {}
    try:
        with open(txt_path, 'r') as f:
            for line in f:
                if ':' in line:
                    key, value = line.split(':', 1)
                    key = key.strip()
                    value = value.strip()
                    if value:
                        parameters[key] = value
        return parameters
    except Exception as e:
        print(f"Error extracting parameters from {txt_path}: {e}")
        return None

def extract_actions_parameters(json_path):
    try:
        with open(json_path, 'r') as f:
            data = json.load(f)
            insole_length = data.get('InsoleLength', {}).get('LengthInMM', None)
            insole_height = data.get('InsoleHeight', {}).get('Height', None)
            insole_arch_height = data.get('InsoleArch', {}).get('Height', None)
            smooth_insole_height = data.get('SmoothInsole', {}).get('SmoothFactor', None)
            return {
                'InsoleLength': insole_length,
                'InsoleHeight': insole_height,
                'InsoleArchHeight': insole_arch_height,
                'SmoothInsoleHeight': smooth_insole_height
            }
    except Exception as e:
        print(f"Error reading {json_path}: {e}")
        return None

base_dir = r'/Users/elvisechefu/Desktop/LutraCAD2/ai_skdfjsdfJDIKFJsihfdsf8472983hdsof'
data = []

for subfolder in os.listdir(base_dir):
    subfolder_path = os.path.join(base_dir, subfolder)
    if os.path.isdir(subfolder_path):
        foot_scans = [f for f in os.listdir(subfolder_path) if f.endswith('.stl') and not f.endswith('library.stl')]
        markers_txt_files = [f for f in os.listdir(subfolder_path) if f.endswith('markers.txt')]

        # Look for the actions_filtered.dat file in the current subfolder
        actions_file = None
        for root, dirs, files in os.walk(subfolder_path):
            for file in files:
                if file.endswith('actions_filtered.dat'):
                    actions_file = os.path.join(root, file)
                    break
            if actions_file:
                break

        if not markers_txt_files or not actions_file:
            continue

        for foot_scan in foot_scans:
            foot_scan_path = os.path.join(subfolder_path, foot_scan)

            markers_params = {}
            for txt_file in markers_txt_files:
                txt_path = os.path.join(subfolder_path, txt_file)
                markers_params = extract_markers_parameters(txt_path)
                if markers_params:
                    break

            if markers_params is None:
                continue

            actions_params = extract_actions_parameters(actions_file)

            # Check if any marker parameters exist before appending to data
            if any(markers_params.values()):
                data.append({
                    'Subfolder': subfolder,
                    'Foot Scan': foot_scan,
                    'MTP1': markers_params.get('MTP1', None),
                    'MTP5': markers_params.get('MTP5', None),
                    'HEEL': markers_params.get('HEEL', None),
                    'ARCH': markers_params.get('ARCH', None),
                    'LENGTH': markers_params.get('LENGTH', None),
                    'WIDTH': markers_params.get('WIDTH', None),
                    'InsoleLength': actions_params.get('InsoleLength', None),
                    'InsoleHeight': actions_params.get('InsoleHeight', None),
                    'InsoleArchHeight': actions_params.get('InsoleArchHeight', None),
                    'SmoothInsoleHeight': actions_params.get('SmoothInsoleHeight', None)
                })

df = pd.DataFrame(data)
df.to_csv('parameters_actions.csv', index=False)

# Splitting the DataFrame into 10 batches
num_batches = 13
batch_size = len(df) // num_batches

for batch_num in range(num_batches):
    start_index = batch_num * batch_size
    if batch_num == num_batches - 1:  # handle the last batch to include all remaining rows
        end_index = len(df)
    else:
        end_index = (batch_num + 1) * batch_size

    df_batch = df.iloc[start_index:end_index]
    batch_csv_path = f'batch_{batch_num + 1}.csv'
    df_batch.to_csv(batch_csv_path, index=False)

print("Data has been split into 10 batches and saved to CSV files.")

#BATCH 1
# Load batch_8.csv
batch_df = pd.read_csv('batch_1.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_1_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 2
# Load batch_8.csv
batch_df = pd.read_csv('batch_2.csv')


# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_2_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 3

# Load batch_8.csv
batch_df = pd.read_csv('batch_3.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_3_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 4

# Load batch_8.csv
batch_df = pd.read_csv('batch_4.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_4_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 5

# Load batch_8.csv
batch_df = pd.read_csv('batch_5.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_5_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 6

# Load batch_8.csv
batch_df = pd.read_csv('batch_6.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_6_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 7

# Load batch_8.csv
batch_df = pd.read_csv('batch_7.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_7_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 8

# Load batch_8.csv
batch_df = pd.read_csv('batch_8.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_8_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 9

# Load batch_8.csv
batch_df = pd.read_csv('batch_9.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_9_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 10

# Load batch_8.csv
batch_df = pd.read_csv('batch_10.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_10_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 11

# Load batch_8.csv
batch_df = pd.read_csv('batch_11.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_11_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 12

# Load batch_8.csv
batch_df = pd.read_csv('batch_12.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_12_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

#BATCH 13

# Load batch_8.csv
batch_df = pd.read_csv('batch_13.csv')

# Ensure the output directory for screenshots exists relative to CSV
output_screenshots_dir = 'screenshots'
os.makedirs(output_screenshots_dir, exist_ok=True)

# Create a new directory for insole screenshots
insole_screenshots_dir = 'insole_screenshots'
os.makedirs(insole_screenshots_dir, exist_ok=True)

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# Iterate over each row in the batch DataFrame
screenshot_paths = []
all_insole_paths = []
all_insole_screenshot_paths = []
for index, row in batch_df.iterrows():
    subfolder = row['Subfolder']
    foot_scan = row['Foot Scan']
    
    subfolder_path = os.path.join(base_dir, subfolder)
    foot_scan_path = os.path.join(subfolder_path, foot_scan)
    insole_paths = extract_insoles(subfolder_path)
    
    # Foot scan screenshot
    screenshot_filename = f"{subfolder}_3D_foot_scan.png"
    screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
    capture_screenshot(foot_scan_path, screenshot_path)
    screenshot_paths.append(screenshot_path)
    
    # Insole screenshots
    insole_screenshot_paths_per_row = []
    for i, insole_path in enumerate(insole_paths):
        insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
        insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
        capture_screenshot(insole_path, insole_screenshot_path)
        insole_screenshot_paths_per_row.append(insole_screenshot_path)
    
    all_insole_paths.append('|'.join(insole_paths))
    all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

# Add the screenshot paths and insole paths to the DataFrame
batch_df['3D Foot Scan'] = screenshot_paths
batch_df['Insole'] = all_insole_paths
batch_df['Insole Screenshot'] = all_insole_screenshot_paths

# Save the DataFrame and images to HDF5
with h5py.File('batch_13_final.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in batch_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = batch_df.to_json()

# Function to capture and save a screenshot of an STL file
def capture_screenshot(stl_path, output_path):
    if os.path.exists(output_path):
        return  # If screenshot already exists, return early
    try:
        mesh = pv.read(stl_path)
        plotter = pv.Plotter(off_screen=True)
        plotter.add_mesh(mesh)
        plotter.show(auto_close=False)
        plotter.screenshot(output_path, transparent_background=True)
        plotter.close()
    except Exception as e:
        print(f"Error visualizing mesh: {e}")

# Function to extract all insoles from subfolders
def extract_insoles(subfolder_path):
    insole_paths = []
    for file in os.listdir(subfolder_path):
        if file.endswith("library.stl"):
            insole_path = os.path.join(subfolder_path, file)
            insole_paths.append(insole_path)
    return insole_paths

# List to hold all batch DataFrames
all_batches_df = []

#COMBINING ALL THE BATCHES 

# List of batch files
batch_files = [f'batch_{i}_final.h5' for i in range(1, 14)]

for batch_file in batch_files:
    with h5py.File(batch_file, 'r') as hf:
        # Read the DataFrame stored as JSON in the HDF5 file
        batch_df = pd.read_json(hf.attrs['DataFrame'])
        
        # Iterate over each row in the batch DataFrame
        screenshot_paths = []
        all_insole_paths = []
        all_insole_screenshot_paths = []
        for index, row in batch_df.iterrows():
            subfolder = row['Subfolder']
            foot_scan = row['Foot Scan']
            
            subfolder_path = os.path.join(base_dir, subfolder)
            foot_scan_path = os.path.join(subfolder_path, foot_scan)
            insole_paths = extract_insoles(subfolder_path)
            
            # Foot scan screenshot
            screenshot_filename = f"{subfolder}_3D_foot_scan.png"
            screenshot_path = os.path.join(output_screenshots_dir, screenshot_filename)
            capture_screenshot(foot_scan_path, screenshot_path)
            screenshot_paths.append(screenshot_path)
            
            # Insole screenshots
            insole_screenshot_paths_per_row = []
            for i, insole_path in enumerate(insole_paths):
                insole_screenshot_filename = f"{subfolder}_insole_{i}.png"
                insole_screenshot_path = os.path.join(insole_screenshots_dir, insole_screenshot_filename)
                capture_screenshot(insole_path, insole_screenshot_path)
                insole_screenshot_paths_per_row.append(insole_screenshot_path)
            
            all_insole_paths.append('|'.join(insole_paths))
            all_insole_screenshot_paths.append('|'.join(insole_screenshot_paths_per_row))

        # Add the screenshot paths and insole paths to the DataFrame
        batch_df['3D Foot Scan'] = screenshot_paths
        batch_df['Insole'] = all_insole_paths
        batch_df['Insole Screenshot'] = all_insole_screenshot_paths
        
        # Append the processed batch DataFrame to the list
        all_batches_df.append(batch_df)

# Concatenate all batch DataFrames into one
combined_df = pd.concat(all_batches_df, ignore_index=True)

# Save the combined DataFrame and images to a new HDF5 file
with h5py.File('combined_batches.h5', 'w') as hf:
    # Convert images to base64 and store in HDF5 dataset
    for index, row in combined_df.iterrows():
        # Handle 3D Foot Scan image
        image_path = row['3D Foot Scan']
        if pd.notnull(image_path):
            with open(image_path, "rb") as img_file:
                encoded_img = base64.b64encode(img_file.read()).decode('utf-8')
                hf.create_dataset(f'{index}/3D_Foot_Scan', data=encoded_img)
        else:
            hf.create_dataset(f'{index}/3D_Foot_Scan', data='')
        
        # Handle Insole images
        insole_paths = row['Insole'].split('|')
        for i, insole_path in enumerate(insole_paths):
            if pd.notnull(insole_path) and os.path.exists(insole_path):
                with open(insole_path, "rb") as insole_file:
                    encoded_insole = base64.b64encode(insole_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_{i}', data=encoded_insole)
            else:
                hf.create_dataset(f'{index}/Insole_{i}', data='')
        
        # Handle Insole Screenshot images
        insole_screenshot_paths = row['Insole Screenshot'].split('|')
        for i, insole_screenshot_path in enumerate(insole_screenshot_paths):
            if pd.notnull(insole_screenshot_path) and os.path.exists(insole_screenshot_path):
                with open(insole_screenshot_path, "rb") as insole_screenshot_file:
                    encoded_insole_screenshot = base64.b64encode(insole_screenshot_file.read()).decode('utf-8')
                    hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data=encoded_insole_screenshot)
            else:
                hf.create_dataset(f'{index}/Insole_Screenshot_{i}', data='')

    # Store DataFrame as JSON within HDF5
    hf.attrs['DataFrame'] = combined_df.to_json()

# Display the DataFrame with HTML formatting in Jupyter Notebook
from IPython.display import display, HTML
display(HTML(combined_df.to_html(escape=False, formatters={
    '3D Foot Scan': lambda x: f'<img src="{x}" width="100"/>' if x else '',
    'Insole Screenshot': lambda x: ' '.join([f'<img src="{path}" width="100"/>' for path in x.split("|") if path]) if x else ''
})))
# DATA SPLITTING, PREPROCESSING AND TRAINING
```
